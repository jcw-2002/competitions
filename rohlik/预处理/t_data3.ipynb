```python
import pandas as pd
# X_val = pd.read_csv('x_val_date.csv').sort_values(by=['unique_id', 'date'])
# 读取数据
X_train = pd.read_csv('train_d.csv')
```

```python
X_train
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = X_train
# 将 'date' 列转换为日期类型
df['date'] = pd.to_datetime(df['date'])

# 按 'warehouse' 和 'date' 分组，计算 sales 的平均值
grouped_df = df.groupby(['warehouse', 'date'])['sales'].mean().reset_index()

# 绘制折线图
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=grouped_df,
    x='date',
    y='sales',
    hue='warehouse',
    marker='o',  # 设置端点样式为圆圈
    markersize=4,  # 设置端点大小
    linewidth=1  # 设置线条粗细
)
plt.title('Average Sales by Warehouse Over Time')
plt.xlabel('Date')
plt.ylabel('Average Sales')
plt.legend(title='Warehouse')
plt.grid(True)
plt.show()
```

```python
# 按 'date' 分组，计算 total_orders 和 sales 的平均值
grouped_df = df.groupby('date')[['total_orders', 'sales']].mean().reset_index()

# 将 total_orders 的值除以 100
grouped_df['total_orders'] = grouped_df['total_orders'] / 100

# 绘制散点图
plt.figure(figsize=(12, 6))

# 绘制 total_orders 的散点图
sns.scatterplot(data=grouped_df, x='date', y='total_orders', color='blue', label='Average Total Orders (divided by 100)', s=50)

# 绘制 sales 的散点图
sns.scatterplot(data=grouped_df, x='date', y='sales', color='red', label='Average Sales', s=50)

# 添加标题和标签
plt.title('Average Total Orders and Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Average Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 按 'date' 分组，计算 sales 和 sell_price_main 的平均值
grouped_df = df.groupby('date')[['sales', 'sell_price_main']].mean().reset_index()

# 绘制散点图
plt.figure(figsize=(12, 6))

# 绘制 sales 的散点图
sns.scatterplot(data=grouped_df, x='date', y='sales', color='blue', label='Average Sales', s=10)

# 绘制 sell_price_main 的散点图
sns.scatterplot(data=grouped_df, x='date', y='sell_price_main', color='red', label='Average Sell Price (Main)', s=10)

# 添加标题和标签
plt.title('Average Sales and Sell Price (Main) Over Time')
plt.xlabel('Date')
plt.ylabel('Average Value')
plt.legend()
plt.grid(True)
plt.show()
```

## 处理1

```python
# 定义折扣列
discount_columns = [
    'type_0_discount', 'type_1_discount', 'type_2_discount',
    'type_3_discount', 'type_4_discount', 'type_5_discount', 'type_6_discount'
]

# 1. 是否有折扣列
df['is_discount'] = df[discount_columns].apply(lambda row: 1 if row.any() else 0, axis=1).astype(bool)

# 2. 最高折扣值列
df['max_discount'] = df[discount_columns].apply(lambda row: row.max(), axis=1).astype('Float64')
```

```python
df['is_discount'].value_counts()
```

```python
df.drop(columns=discount_columns, inplace=True)
```

## product_id and L_name


```python
import pandas as pd
# 读取CSV文件
train = pd.read_csv('sales_train.csv')
inventory = pd.read_csv('inventory.csv')
```

```python
train = train.dropna().drop(columns=['availability'])# 测试集中没有availability列，所以删除
```

```python
train = pd.merge(train, inventory, on=['unique_id', 'warehouse'], how='left')
```

```python
train.columns
```

```python
train['product_unique_id'].value_counts()
```

```python
train['sales'].max()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# 按 product_unique_id 分组，计算 sales 的均值
product_sales_mean = train.groupby('product_unique_id')['sales'].mean().reset_index()

# 按 sales 均值从高到低排序
product_sales_mean = product_sales_mean.sort_values(by='sales', ascending=False)

# 绘制柱状图
plt.figure(figsize=(10, 6))
sns.barplot(data=product_sales_mean, x='product_unique_id', y='sales', palette='viridis', order=product_sales_mean['product_unique_id'])
plt.title('Average Sales by Product Unique ID')
plt.xlabel('Product Unique ID')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)  # 旋转 x 轴标签，避免重叠
plt.grid(True, axis='y')  # 添加水平网格线

# 设置纵坐标范围自适应
plt.ylim(0, product_sales_mean['sales'].max() * 1.1)  # 纵坐标上限为最大值的 1.1 倍

plt.show()
```

## log平滑替换product_id

```python
import  numpy as np
# 对 sales 均值进行对数变换
product_sales_mean['log_sales'] = np.log1p(product_sales_mean['sales'])  # 使用 log1p 避免对 0 取对数

# 将对数变换后的值替换原始的 sales 均值
product_sales_mean['sales'] = product_sales_mean['log_sales']
```

```python
# 绘制对数变换前后的对比图
plt.figure(figsize=(12, 6))
sns.barplot(data=product_sales_mean, x='product_unique_id', y='sales', palette='viridis')
plt.title('Log-Transformed Average Sales by Product Unique ID')
plt.xlabel('Product Unique ID')
plt.ylabel('Log-Transformed Average Sales')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```

```python

# 删除临时列
product_sales_mean.drop(columns=['log_sales'], inplace=True)

# 将变换后的值映射回原始数据（如果需要）
train = train.merge(product_sales_mean, on='product_unique_id', suffixes=('', '_log'))
```
