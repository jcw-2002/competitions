```python
train.head()
```

```python
# 将 'date' 列转换为 datetime 类型
import numpy as np
for df in [train, test]:
    df['date'] = pd.to_datetime(df['date'])
    # 提取年、月、日
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['day'] = df['date'].dt.day

    # 对月份进行周期编码
    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

    # 对日进行周期编码
    df['day_sin'] = np.sin(2 * np.pi * df['day'] / 31)
    df['day_cos'] = np.cos(2 * np.pi * df['day'] / 31)
    df.drop(columns=['month', 'day'], inplace=True)
```

```python
import matplotlib.pyplot as plt

# 生成所有月份
months = np.arange(1, 13)
month_sin = np.sin(2 * np.pi * months / 12)
month_cos = np.cos(2 * np.pi * months / 12)

# 绘制月份的正弦和余弦值
plt.figure(figsize=(10, 4))
plt.plot(months, month_sin, label='month_sin')
plt.plot(months, month_cos, label='month_cos')
plt.xlabel('Month')
plt.ylabel('Value')
plt.title('Cyclic Encoding for Month')
plt.legend()
plt.show()
```

```python
# 生成所有日
days = np.arange(1, 32)
day_sin = np.sin(2 * np.pi * days / 31)
day_cos = np.cos(2 * np.pi * days / 31)

# 绘制日的正弦和余弦值
plt.figure(figsize=(10, 4))
plt.plot(days, day_sin, label='day_sin')
plt.plot(days, day_cos, label='day_cos')
plt.xlabel('Day')
plt.ylabel('Value')
plt.title('Cyclic Encoding for Day')
plt.legend()
plt.show()
```

```python
#使用 groupby 按年份分组，并计算每一年里的平均销售额：
yearly_avg_sales = train.groupby('year')['sales'].mean().reset_index()
yearly_avg_sales.columns = ['year', 'avg_sales']  # 重命名列
yearly_avg_sales
```

```python
# 感觉并没有随着年份变化销售额会有所变化（20-23，很低，疫情影响吧？）
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(yearly_avg_sales['year'], yearly_avg_sales['avg_sales'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Average Sales')
plt.title('Average Sales by Year')
plt.grid(True)
plt.show()
```
