```python
import numpy as np
import pandas as pd
from sklearn.utils import resample
```

```python
error_list=['物料推送装置故障1001',
'物料检测装置故障2001',
'填装装置检测故障4001',
'填装装置定位故障4002',
'填装装置填装故障4003',
'加盖装置定位故障5001',
'加盖装置加盖故障5002',
'拧盖装置定位故障6001',
'拧盖装置拧盖故障6002'
]
error_value=[1001, 2001, 4001, 4002, 4003, 5001, 5002, 6001, 6002]
```

```python
for i in range(1,11):
    name = 'M'+ str(100+i)
    df_name= './data/附件1/'+ name + '.csv'
    df = pd.read_csv(df_name)
    print(df_name)
    for m in range(len(error_list)):
        # 假设df是你的大dataframe，'类别列'是你的目标列
        # 首先，分别获取多数类和少数类的样本
        df_majority = df[df[error_list[m]] == 0]
        df_minority = df[df[error_list[m]] == int(error_value[m])]

        # 然后进行下采样
        df_majority_downsampled = resample(df_majority, 
                                           replace=False,    # 不进行替换，意味着不会有重复
                                           n_samples=len(df_minority),     # 少数类的数量
                                           random_state=123) # 随机种子保证每次结果一致

        # 最后，合并少数类和下采样后的多数类样本
        df_balanced = pd.concat([df_minority, df_majority_downsampled])
        csv_name = './data/'+ name +  str(error_value[m]) + '.csv'
        df_balanced.to_csv(csv_name)
        # 检查新的类别分布
        print(df_balanced[error_list[m]].value_counts())
    print()
```

```python
df_unerror = df[~df.index.isin(df_error.index)]
df_unerror
```

```python
from sklearn.utils import resample

# 假设df_unerror是多数类样本的dataframe，df_error是少数类样本的dataframe

# 进行下采样，从多数类样本中随机选择与少数类样本相同数量的样本
df_unerror_downsampled = resample(df_unerror, 
                                  replace=False,    # 不进行替换，意味着不会有重复
                                  n_samples=len(df_error),     # 少数类样本的数量
                                  random_state=123) # 随机种子保证每次结果一致

# 合并少数类样本和下采样后的多数类样本
df_balanced = pd.concat([df_error, df_unerror_downsampled])

# 检查新的类别分布
print(df_balanced['物料推送装置故障1001'].value_counts())
```
