```python

# 源文件夹路径列表
source_dirs = ['87', '131', '318','train']
# 目标文件夹路径
dir = 'add_json/'
# 遍历每个源文件夹，读取并解析对应的 JSON 文件
all_images = []
all_annotations = []
for source in source_dirs:
    # 构建 JSON 文件路径
    json_file = os.path.join(dir, f'{source}.json')
    with open(json_file, 'r') as f:
        data = json.load(f)

    images_infos = data['images']
    # 获取该图像的标注信息
    annotations = data['annotations']

    # 获取所有类别信息
    categories = data['categories']
    all_annotations.extend(annotations)

    all_images.extend(images_infos)

```

```python
# 创建增强后的JSON数据
augmented_json_data = {
    'info': info,
    'images': all_images,
    'annotations': all_annotations,
    'categories': categories
}
augmented_json_path = os.path.join(dir, f"all_json.json")
with open(augmented_json_path, 'w') as json_file:
    json.dump(augmented_json_data, json_file, indent=4)
```
